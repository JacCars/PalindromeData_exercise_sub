### Questions

#### What is the total number of people living with HIV (NoPLHIV) in the listed districts according to the Survey estimate?

Answer: 6409903

```python
hiv_df[hiv_df['Estimate'] == 'Survey']['NoPLHIV'].sum()
```

#### What is the average NoPLHIV of the two estimates used for “Xhariep”?
Answer : 12247.5

```python
hiv_df[hiv_df['District'] == 'Xhariep']['NoPLHIV'].mean()
```

#### Add a column and populate it with the number of people not living with HIV for each row. 
Assumption: 
Answer : See 'NoP_not_LHIV' column in hiv_df below

```python
hiv_df['NoP_not_LHIV'] = (hiv_df['NoPLHIV']/hiv_df['Prevalence_%']*100 - hiv_df['NoPLHIV']).astype(int)
```

```python
hiv_df
```

#### What is the total NoPLHIV in all the cities (districts with the word “city” or “metro” in the name)? 
Answer:
* By Survey - 2466420 people
* By FH model - 2475274 people

```python
# for Survey:
print(hiv_df[hiv_df['District'].str.contains('city', case=False) | hiv_df['District'].str.contains('metro', case=False) & (hiv_df['Estimate'] == 'Survey')]['NoPLHIV'].sum())

# for FH model:
print(hiv_df[hiv_df['District'].str.contains('city', case=False) | hiv_df['District'].str.contains('metro', case=False) & (hiv_df['Estimate'] == 'Fay-Heriott')]['NoPLHIV'].sum())
```
